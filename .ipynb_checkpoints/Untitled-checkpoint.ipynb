```python
def calculate_block_areas(json_data):
    arr = []
    fullTextAnnotation_pages = json_data['fullTextAnnotation']['pages']

    for page in fullTextAnnotation_pages:
        blocks = page['blocks']
        for block in blocks:
            block_vertices = block['boundingBox']['vertices']

            # Find x coordinates
            x_values = [vertex['x'] for vertex in block_vertices]
            max_x = max(x_values)
            min_x = min(x_values)

            # Find y coordinates
            y_values = [vertex['y'] for vertex in block_vertices]
            max_y = max(y_values)
            min_y = min(y_values)

            block_x = max_x - min_x
            block_y = max_y - min_y

            area = block_x * block_y
            arr.append(area)

    return arr
```
