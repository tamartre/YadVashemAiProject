```python
#get JSON and search word!
import base64
import requests
import json
import io
import ast
import numpy as np
import glob

url = "https://vision.googleapis.com/v1/images:annotate"
querystring = {"key": "AIzaSyBmcYM1GuYXf-ZPs5ppdFgLFSwQ5-LB33I"}

def img_req(path):
    headers = {
        'Content-Type': 'application/json'
    }

    with open(path, 'rb') as image_file:
        encoded_image = base64.b64encode(image_file.read()).decode('utf-8')

    payload = {
        "requests": [{
            "image": {
                "content": encoded_image
            },
            "features": [{
                "type": "TEXT_DETECTION"
            }]
        }]
    }

    response = requests.post(url, json=payload, headers=headers, params=querystring)

    return response.text

def search_word_and_get_position(text, target_word):
    response_data = []
    
    if target_word.lower() in text.lower():
        dict_text = ast.literal_eval(text)
        detected_texts = dict_text["responses"][0]["textAnnotations"]
        
        for annotation in detected_texts:
            description = annotation["description"].lower()
            
            words = description.split()
            for i, word in enumerate(words):
                # Check if the word matches the target word and is a standalone word
                if word == target_word.lower() and (i == 0 or not words[i-1].isalpha()) and (i == len(words)-1 or not words[i+1].isalpha()):
                    vertices = annotation["boundingPoly"]["vertices"]
                
                    left = vertices[0]["x"]
                    top = vertices[0]["y"]
                    right = vertices[1]["x"]
                    bottom = vertices[2]["y"]
                    
                    width = right - left
                    height = bottom - top

                    response_data.append({
                        "word": target_word,
                        "left": left,
                        "top": top,
                        "right": right,  # Include the "right" position
                        "width": width,
                        "height": height
                    })
                    
                    break  # Stop searching for the target word in this description
    
    return response_data


folder_name = '123'
target_words = ["Moshe","Cohen"]

all_texts = ''

for path in glob.glob(folder_name + "/*.png"):
    print(path)
    text = img_req(path)
    dict_text = ast.literal_eval(text)
    
    if "responses" in dict_text and dict_text["responses"]:
        detected_text = dict_text["responses"][0]["textAnnotations"][0]["description"]
        all_texts += detected_text + "\n"
        
        # Search for each target word and get their positions
        for target_word in target_words:
            search_results = search_word_and_get_position(text, target_word)
            
            for result in search_results:
                left = result["left"]
                top = result["top"]
                right = left + result["width"]  # Calculate 'right' based on 'left' and 'width'
                width = result["width"]
                height = result["height"]
                print(f"Found '{result['word']}' at position: Left={left}, Top={top}, Right={right}, Width={width}, Height={height} in {path}")

all_texts = all_texts.encode('utf-8')

with open(folder_name + '.txt', 'w', encoding='utf-8') as f:
    f.write(all_texts.decode('utf-8'))
```
